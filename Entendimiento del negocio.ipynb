ENTENDIMIENTO DEL NEGOCIO

Determinando los objetivos del negocio (Overview)

La Financiera de Desarrollo Nacional (FDN) es un banco de desarrollo único en Colombia especializado en la financiación y estructuración de proyectos de infraestructura. Se enfoca en superar vacíos del mercado con el fin de movilizar fuentes de financiación que permitan el desarrollo de la infraestructura. Es un Establecimiento de Crédito y en esa medida está vigilada por la Superintendencia Financiera de Colombia. 

Recientemente ha estado en la búsqueda de líneas de negocio que le permitan diversificar el riesgo dado que actualmente se encuentra concentrada en la financiación a carreteras 4G. Dentro de las líneas de negocio que ha considerado se encuentra la línea llamada Underwriting. Este producto busca comprometer a una entidad estructuradora (FDN) con el remanente de los títulos no colocados en una emisión por parte de una empresa (entidad emisora de títulos en el mercado de capitales). Con el propósito de sacar este producto al mercado, se ha diseñado una metodología para determinar el Underwriting fee que debe cobrar la FDn a sus cliente. La metodología es liderada por el equipo de Desarrollo de Producto de la entidad; sin embargo involucra diferentes áreas; entre ella las de Riesgos Financieros. Particularmente en ésta área, el interés radica en determinar el valor que debería aportar a la tarifa el riesgo de mercado, liquidez y crédito que asume la FDN por mantener el remanente de una colocación en el portafolio de Posición Propia. 

El resultado del ejercicio será exitoso si se logra, para cualquier emisión estudiada, determinar cuánto le corresponde agregar a la tarifa por el riesgo de mercado, liquidez y crédito que asume  la FDN al incorporar el remanente de dicha emisión en su portafolio, basado en sus características faciales.

•	Definiendo los objetivos del negocio
El problema de determinar la tarifa del underwriting está a cargo del área de Desarrollo de Producto; sin embargo el área objetivo de este proyecto es el área de Riesgo Financiero dado que lo que se pretende es identificar, cuántos puntos básicos se deben agregar a la tarifa en función del riesgo de mercado, liquidez y crédito que aportarían los títulos sujetos a Underwriting.

La motivación del proyecto es poder colocarle precio al riesgo de mercado, liquidez y crédito con el fin de completar el valor de la tarifa para el Underwriting previo a lanzar este producto al mercado. En la medida en que sea una tarifa competitiva la FDN podrá diversificar su negocio y promover la infraestructura del país, lo cual es primordial en la definición de su misión.

El proyecto actualmente no utiliza Datamining; sin embargo utiliza metodologías de valoración independientes para el riesgo de mercado, liquidez y crédito. Se quiere revisar si los valores arrojados por dichas metodologías pueden ser entendidas de mejor manera bajo el proceso de datamining, lo anterior dado que el movimiento en tasa de los títulos en el mercado de valores recoge los tres efectos y no es trivial resolver cuánto aporta cada tipo de riesgo.

La solución actual fue acogida dentro del equipo de trabajo y es la propuesta que se está trabajando para el lanzamiento del producto. Sin embargo, en la medida en la que se logre entender mejor cuánto riesgo asume la entidad en cada uno de los tres frentes (riesgo de mercado, liquidez y crédito), se podrá calibrar la tarifa de mejor manera y se podrá ajustar a condiciones de mercado.

•	Criterios de éxito del negocio
Lograr determinar el porcentaje de riesgo de mercado, liquidez y crédito que representa los movimientos (en tasa) de un título con unas características faciales específicas en un periodo de tiempo específico.


EVALUANDO LA SITUACIÓN
•	Inventario de fuentes
La fuente de información principal serán series de precios de bonos de infraestructura (Bloomberg o proveedor de precios), o con características faciales smilares a la del título evaluado, tiempo estimado de liquidación del título (entregado por el equipo de la Tesorería), serie histórica de TES (títulos emitidos por el Gobierno Nacional). 

La compañía tiene acceso a la información y puede descargarla en sus servidores; sin embargo la información es de uso confidencial. Respecto a consultas con el equipo de trabajo, hay disponibilidad para agendar sesiones de trabajo con el fin de despejar dudas particulares.


•	Requerimientos supuestos y constantes
La información puede ser recopilada directamente desde Bloomberg o el proveedor de precios y no requiere involucrar personal de otras áreas.

•	Riesgos y contingencias
-Que el resultado obtenido refleje un nivel de tarifa tal que no se pueda ofrecer la línea de negocio pensada. Esto podría generar un impacto en la metodología global de cálculo de tarifa con el fin de buscar ajustes en otros equipos de trabajo para llegar a un nivel competitivo.
-El tiempo que se tome en desarrollar este ejercicio podría dejar a la FDN por fuera de algunos negocios, que en este contexto, a veces salen de manera inesperada y dan poco tiempo de reacción.

•	Terminología
Underwriting - Producto que busca comprometer a una entidad estructuradora con el remanente de los títulos no colocados en una emisión por parte de una empresa (entidad emisora de títulos en el mercado de capitales).
Puntos básicos (PBS) - 100 puntos básicos corresponden a un 1%.

•	Análisis costo/beneficio
Lograr valorar el riesgo de mercado, liquidez y crédito permite a la FDN entender las variables o la relación que tienen los títulos con el mercado de tal manera que ante una situación adversa de mercado se puede calibrar la tarifa.

El desarrollo de este proyecto no requirirá costos adicionales salvo que la información que se descargue sea tal que sobrepase los límites permitidos por Bloomberg, en cuyo caso se deberá esperar el cierre de mes para que asignen nuevamente la posibilidad de descargar datos.

Como beneficio también se podría concluir que se debe incurrir en la revisión de las demás variables, bajo esta misma metodología, con el fin de entender mejor las variables que impactan el precio.


DETERMINANDO LAS METAS DEL EJERCICIO DE MINERÍA DE DATOS
•	Metas del data-mining
-Usar información histórica de precios de bonos de infraestructura locales e internacionales de países emergentes similares a Colombia, para evaluar si hay una correlación con los títulos del Gobierno de cada país.
-Evaluar el punto anterior a la luz de la calificación emitida por una Sociedad Calificadora de Valores, plazo y tasa de indexación de los títulos.
-Establecer un criterio que discrimine cuánto de los movimientos de un título corresponde a riesgo de mercado, liquidez y crédito.
-Definir cómo se debe agregar el resultado al valor de la tarifa.

•	Criterios de éxito del data-mining
-La evaluación del dataminig será exitosa si se logran explicar los movimientos en la tarifa por concepto de riesgo de mercado, liquidez y crédito ante distintas situaciones adversas de mercado donde se estrese alguno de los factores de la economía o antes situaciones adversas particulares para el emisor del título.
-El ejercicio debe arrojar una segmentación según plazo, calificación o indexación. No podrá ser un porcentaje fijo.
